```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('2018.xls',dtype={'社保卡号':str})
df.head(5)
```

```python
df[df.isna().any(axis=1)] # 查看所有有缺失值的行
```

```python
df.rename({'购药时间':'销售时间'},axis=1,inplace=True)
df.dropna(subset=['销售时间','社保卡号'],how='any',inplace=True) # 删除指定列的空值
df.reset_index(drop=True,inplace=True)

df.info()
```

```python
df['星期'] = df['销售时间'].str[11:]
df['销售时间'] = df['销售时间'].str[0:10]
```

```python
df['销售时间'] = pd.to_datetime(df['销售时间'], errors='coerce') # 日期出错，返回nan
df.dropna(how='any',inplace=True)
df.reset_index(drop=True,inplace=True)
df.info()
```

```python
df['商品编码'] = df['商品编码'].astype('str')
df.info()
```

```python
# df[df['销售数量']<0]
df = df[df['销售数量']>0]
df.reset_index(drop=True,inplace=True)
df.describe()
df.info()
```

```python
## 删除重复值
df.drop_duplicates(inplace=True)
df
```

```python
df.sort_values('销售时间',inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

# 月均消费次数

```python
num = df['销售数量'].count()
start = df.iloc[0]['销售时间']
start
end = df.iloc[-1]['销售时间']
end
days = end-start
days
months = round(days/pd.Timedelta(days = 30),0)
months
count_months = num /months
count_months
num
print(f'共计消费{num}次')
print(f'一共{months}月')
print(f'月均消费{round(count_months,0)}次')
```

# 月均消费金额

```python
money = sum(df['实收金额'])
money
money_months = money/months
print(f'月收入金额{round(money_months,2)}')
```
